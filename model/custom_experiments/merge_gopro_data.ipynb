```python
%matplotlib inline
import os
import os.path
import pandas as pd
import numpy as np

dirs = ['/data/neon/gopro/pro', '/data/neon/gopro/biking/pro', '/data/neon/gopro/party',
        '/data/neon/gopro/snowsports/pro']
```

```python
def iterate_pickles(d):
    for fn in os.listdir(d):
        if fn.endswith('.pkl'):
            data = pd.read_pickle(os.path.join(d, fn))
            yield fn, data.loc[:,150:]
```

```python
sample = pd.read_pickle(os.path.join(dirs[1], '8FIeRgonxoI.pkl'))
sample.loc[:,150:]
```

```python
dfs = {}
for d in dirs:
    fns, vid_samples = zip(*iterate_pickles(d))
    dfs[d] = pd.concat(vid_samples, axis=1, keys=fns)
    
```

```python
big_df = pd.concat(dfs.itervalues(), axis=1, keys=dfs.iterkeys())
big_df
```

```python
big_df.to_pickle('/data/neon/gopro/many_pro.pkl')
```
